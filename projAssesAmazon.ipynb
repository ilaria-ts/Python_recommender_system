### Recommendation Model: Some of the movies hadnâ€™t been watched and therefore, are not rated by the users.  Netflix would like to take this as an opportunity and build a machine learning recommendation algorithm which provides the ratings for each of the users.


* Divide the data into training and test data
* Build a recommendation model on training data
* Make predictions on the test data

```python
from surprise import Reader
from surprise import accuracy
from surprise import Dataset
from surprise.model_selection import train_test_split
from surprise import SVD
from surprise.model_selection import cross_validate
from surprise.model_selection import GridSearchCV
```

```python
data_melt = data.melt(id_vars=data.columns[0], value_vars=data.columns[1:], var_name = 'movie',value_name = 'ratings')
data_melt
```

```python
data_melt['ratings'].mean()
```

```python
reader = Reader(rating_scale=(1,5))
data_r = Dataset.load_from_df(data_melt.fillna(data_melt['ratings'].mean()), reader=reader)
```

```python
data_r
```

```python
#divide train and test
train,test= train_test_split(data_r, test_size=0.25)
```

```python
alg=SVD()
```

```python
#applying the SVD to the train set
alg.fit(train)
```

```python
pred = alg.test(test)
```

```python
accuracy.rmse(pred)
```

```python
#it seems a faily good approximation

us_id = 'A3R5OBKS7OM2IR'
movie_id = 'Movie1'
r_true = 5.0
alg.predict(us_id,movie_id,r_true, verbose=True)
```

```python
#create a function

def repeat(alg_type, frame, min, max):
    reader = Reader(rating_scale=(1,5))
    
    data_r = Dataset.load_from_df(frame, reader=reader)

    alg=alg_type
    
    print(cross_validate(alg, data_r, measures = ['RMSE','MAE'],cv=5, verbose = True))
    
    print("#"*10)
    
    us_id = 'A3R5OBKS7OM2IR'
    movie_id = 'Movie1'
    r_true = 5.0
    print(alg.predict(us_id,movie_id,r_true, verbose=True))
    
    print("#"*20)
    
    print()
```

```python
#3 different models to compare

repeat(SVD(),data_melt.fillna(data_melt['ratings'].mean()),1,5)
repeat(SVD(),data_melt.fillna(data_melt['ratings'].median()),1,5)
repeat(SVD(),data_melt.fillna(0),1,5)
```

```python
param_grid = {'n_epochs':[20,30],
              'lr_all':[0.005,0.01],
              'n_factors':[50,100] }
```

```python
gs = GridSearchCV(SVD, param_grid, measures=['rmse','mae'],cv=5)
gs.fit(data_r)
```
